```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3장 연습문제
#1-1
((13-10)^2+(19-15)^2)**(1/2)
 
```

```python
#1-2
data = np.array([[13, 19], [10, 17], [12, 15], [16, 12], [18, 16]])
scaler = StandardScaler()
scaler.fit(data)
data_std = scaler.transform(data)
print(data_std)

plt.scatter(data_std[:,0], data_std[:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

rows, columns = data_std.shape

for i in range(rows):
  plt.annotate(i+1, (data_std[i,0], data_std[i,1]))

  kmeans = KMeans(n_clusters=2)
kmeans.fit(data_std)
print(kmeans.labels_)
print(kmeans.cluster_centers_)  
plt.scatter(data_std[:,0], data_std[:,1], c=kmeans.labels_, s=50)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

rows, columns = data_std.shape

for i in range(rows):
  plt.annotate(i+1, (data_std[i,0], data_std[i,1]))  

plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            c='red', s=200, alpha=0.5);
 
for k in range(columns):
  plt.annotate(k, 
              (kmeans.cluster_centers_[k, 0], kmeans.cluster_centers_[k, 1]))
```

```python
#1-2
TWSS = {}
for k in range(1, 6): 
  model = KMeans(n_clusters=k) 
  model.fit(data_std)
  TWSS[k] = model.inertia_

plt.plot(list(TWSS.keys()), list(TWSS.values()), marker='o')
plt.xlabel("Number of Clusters(K)")
plt.ylabel("TWSS")
plt.show()
```
